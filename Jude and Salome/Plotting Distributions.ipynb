```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Plots boxplots for distributions of price and supply risk
mof_score_indexed = pd.read_csv('./mof_score_indexed.csv', sep=',')

ax = plt.subplot(111)
ax = ax.set_ylim(0,20)
plt.title("Distributions of MOF Price and Supply Risk factors")

mof_score_indexed.boxplot(column=['Price', 'Supply risk'])
```

```python
#Plots boxplots of toxicity and radioactivity distributions
mof_score_indexed = pd.read_csv('./mof_score_indexed.csv', sep=',')

ax = plt.subplot(111)
ax = ax.set_ylim(0,1)
plt.title("Distributions of MOF Toxicity and Radioactivity")

mof_score_indexed.boxplot(column=['Toxicity', 'Radioactive'])
```

```python
#Distribution boxplot of MOF densities
mof_score_indexed = pd.read_csv('./mof_score_indexed.csv', sep=',')

ax = plt.subplot(111)
ax = ax.set_ylim(0,4)
plt.title("Distribution of MOF Density")

mof_score_indexed.boxplot(column=['Density'])
```

```python
# Density vs avg CN
mof_score_plot = mof_score_indexed.plot.scatter(x='Average CN',y='Density', s=6)

mof_score_plot.axis(ymin=0,ymax=4)
```
